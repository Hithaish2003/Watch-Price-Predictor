### TypeCasting Feature

```python
df['Price (USD)']=df['Price (USD)'].str.replace(',','')

df['Price (USD)'] = df['Price (USD)'].astype(float)

df['Water Resistance']=df['Water Resistance'].str.replace('meters','')

df['Water Resistance'] = df['Water Resistance'].astype(float)

df.head()

df['Power Reserve']=df['Power Reserve'].str.replace('hours','')

df['Power Reserve']=df['Power Reserve'].str.replace('days','')

df['Power Reserve'] = df['Power Reserve'].astype(float)
```

## Removing Outliers

```python
col=['Case Diameter (mm)','Band Width (mm)','Water Resistance','Price (USD)','Power Reserve']
Q3=df[col].quantile(0.75)
Q1=df[col].quantile(0.25)
IQR=Q3-Q1
lb=Q1-1.5*IQR
ub=Q3+1.5*IQR
outlier=df[((df[col]<=lb) | (df[col]>=ub)).any(axis=1)]
df=df.drop(outlier.index)
```

## MultiLinear Regression

```python
x=df.iloc[:,0:11]
y=df['Price (USD)']
```

```python
from sklearn.feature_selection import SelectKBest,chi2
sh=SelectKBest(chi2,k=3)
sh.fit_transform(x,y)
```

```python
x=df[['Brand','Water Resistance','Power Reserve']]
```

```python
from sklearn.model_selection import train_test_split
xtrainR,xtestR,ytrainR,ytestR=train_test_split(x,y,test_size=0.3,shuffle=True,random_state=1)
```

## Decision Tree Regressor

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,shuffle=True,random_state=1)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor(max_depth= 10,max_features='sqrt', min_samples_leaf= 2, min_samples_split= 5)
dtr.fit(xtrain,ytrain)
```

```python
dtr.score(xtrain,ytrain)
```

```python
dtr.score(xtest,ytest)
```

```python
dtr_ypred=dtr.predict(xtest)
r2_score(ytest,dtr_ypred)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
param_grid = {
    'max_depth': [3, 5, 7, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}
dtr = DecisionTreeRegressor(random_state=42)
grid_search = GridSearchCV(
    estimator=dtr,
    param_grid=param_grid,
    cv=5,
    n_jobs=-1,
    scoring='r2',
    verbose=2
)
grid_search.fit(xtrain, ytrain)
print("Best Parameters:", grid_search.best_params_)
print("Best RÂ² Score:", grid_search.best_score_)
best_model = grid_search.best_estimator_
test_score = best_model.score(xtest, ytest)
print("Test Score:", test_score)
```

```python
import pickle as pk
with open('WatchPrice.pkl','wb')as f:
    pk.dump(grid_search.best_estimator_,f)
```
